## 색깔을 지정하지 않을 경우에 실제 색상을 처리하는 순서

### 확률밀도함수를 하나 정의한다.

```python
def pdf(X, mu, sigma) :
    a = 1. / (sigma * np.sqrt(2.* np.pi))
    b = -1./ (2. * sigma **2)
    return a * np.exp(b * (X -mu)**2)

```

### 실제 들어갈 값을 정의한다.

```python
X = np.linspace(-6,6,1000)
```

###  특정 샘플로 평균만 표준편차를 지정해서 확률밀도함수의 값을 받는다.

```python
aa = pdf(X,0,1)
```
