```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.tri as tri
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
a = np.random.rand(100,2)
```

## 삼각화를 먼저 처리한다

```python
triangles = tri.Triangulation(a[:,0],a[:,1])
```

## 삼각형 그래프를 그린다

```python
plt.triplot(triangles)
```
