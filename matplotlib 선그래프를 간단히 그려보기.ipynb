## 간단한 선그래프 그리기

```python
x = range(100)
```

```python
x
```

```python
y = [ i**2 for i in x]
```

```python
type(y)
```

## x축과 y 축에 넣고 선그래프 그리기

```python
plt.plot(x,y)
```

## show 함수를 가지 사용하면 실제 그래프 객체는 보이지 않는다

```python
plt.plot(x,y)
plt.show()
```
