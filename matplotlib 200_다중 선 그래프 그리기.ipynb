## 특정 함수 내부에 plot 함수를 정의해서 그리기

```python
def plot_slope(x,y) :
    xs = x[1:] - x[:-1]
    ys = y[1:] - y[:-1]
    plt.plot(x[1:],ys/xs)
```

```python
xx = np.linspace(-3,3,100)
```

```python
yy = np.exp(-xx **2)
```

```python
plt.plot(xx,yy)
plot_slope(xx,yy)
plt.show()
```
